## Scipy
We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1. fitting it to the periodic function
2. plot the fit

### Data:

Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25

Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
month = np.arange(1,13)
days = np.linspace(0,12,num=365)
```

```python
def MyFunc(x, avg, a, offset):
    return (avg + a* np.cos((x+offset) *2 * np.pi/ x.max()))
```

```python
p_max, cov_max = optimize.curve_fit(MyFunc, month, max, p0=[40, 20, 0])
p_min, cov_min = optimize.curve_fit(MyFunc, month, min, p0=[20, 10, 0])
```

```python
plt.figure()
plt.plot(month, max, 'ro')
plt.plot(month, min, 'bo')
plt.plot(days, MyFunc(days, *p_max), 'r-', label = 'Fitted for Max', )
plt.plot(days, MyFunc(days, *p_min),  'b-', label = 'Fitted for Min')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
plt.legend(loc='best')
plt.show()
```
